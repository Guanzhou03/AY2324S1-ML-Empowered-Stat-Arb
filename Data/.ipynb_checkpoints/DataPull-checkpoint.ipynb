```python
closedata = pd.read_csv(r'C:/Users/user/Documents/GitHub/AY2324S1-ML-Empowered-Stat-Arb/Data/Feature Data/close.csv').set_index('datadate')

def rsi(data, period = 14, ema = True):
    
    #Returns a pd.Series with the relative strength index.
    close_delta = data.diff()

    # Make two series: one for lower closes and one for higher closes
    up = close_delta.clip(lower=0)
    down = -1 * close_delta.clip(upper=0)
    
    if ema == True:
	    # Use exponential moving average
        ma_up = up.ewm(com = period - 1, adjust=True, min_periods = period).mean()
        ma_down = down.ewm(com = period - 1, adjust=True, min_periods = period).mean()
    else:
        # Use simple moving average
        ma_up = up.rolling(window = period, adjust=False).mean()
        ma_down = down.rolling(window = period, adjust=False).mean()
        
    rsi = ma_up / ma_down
    rsi = 100 - (100/(1 + rsi))
    return rsi

rsidata = pd.DataFrame()

for ticker in closedata.columns:
    data = pd.DataFrame(rsi(closedata.loc[:,ticker]))
    rsidata = pd.concat([rsidata,data],1)
rsidata.to_csv('C:/Users/user/Documents/GitHub/AY2324S1-ML-Empowered-Stat-Arb/Data/Feature Data/rsi.csv')
print("RSI DATA FILE WRITTEN")
```

```python
highdata = pd.read_csv(r'C:/Users/user/Documents/GitHub/AY2324S1-ML-Empowered-Stat-Arb/Data/Feature Data/high.csv').set_index('datadate')
lowdata = pd.read_csv(r'C:/Users/user/Documents/GitHub/AY2324S1-ML-Empowered-Stat-Arb/Data/Feature Data/low.csv').set_index('datadate')
closedata = pd.read_csv(r'C:/Users/user/Documents/GitHub/AY2324S1-ML-Empowered-Stat-Arb/Data/Feature Data/close.csv').set_index('datadate')

def atr(high, low, close, period=14):
    
    tr = np.maximum(high - low, np.abs(high - close.shift()), np.abs(low - close.shift()))
    atr = tr.ewm(alpha=1/period, adjust=False).mean()
    
    return atr

atrdata = pd.DataFrame()

for ticker in closedata.columns:
    data = pd.DataFrame(atr(highdata.loc[:,ticker],lowdata.loc[:,ticker],closedata.loc[:,ticker]))
    atrdata = pd.concat([atrdata,data],1)

atrdata.to_csv('C:/Users/user/Documents/GitHub/AY2324S1-ML-Empowered-Stat-Arb/Data/Feature Data/atr.csv')
print("ATR DATA FILE WRITTEN")
```
